Using Text Similarity:

You can compute a measure of text similarity between posts and use it as a weight for the edges. Techniques such as cosine similarity or Jaccard similarity can be used to quantify the similarity between the textual content of posts.

**Edges and Edges with mapping**

```python
from google.colab import drive
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/reddit_posts_cleaned.csv'
df = pd.read_csv(file_path)

# Calculate text similarity using TF-IDF and cosine similarity
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text_sup_clean'])

# Compute cosine similarity
similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Create a new DataFrame for edges with columns 'source', 'target', and 'weight'
edges_df = pd.DataFrame(columns=['source', 'target', 'weight'])

# Assign values to the new DataFrame based on text similarity
for i in range(len(similarities)):
    for j in range(i + 1, len(similarities)):
        weight = similarities[i, j]
        edges_df = edges_df.append({'source': i, 'target': j, 'weight': weight}, ignore_index=True)

# Save the DataFrame with edges to a new CSV file
output_edges_file_path = '/content/drive/MyDrive/edges.csv'
edges_df.to_csv(output_edges_file_path, index=False)
print(f"Edges DataFrame saved to: {output_edges_file_path}")

# Create a DataFrame to store edges
edges_df = pd.DataFrame(columns=['source', 'target', 'label', 'weight'])

# Populate edges DataFrame
for i in range(len(similarities)):
    for j in range(i + 1, len(similarities)):
        source_label = df.at[i, 'id']
        target_label = df.at[j, 'id']
        similarity_score = similarities[i][j]

        edges_df = edges_df.append({
            'source': i,
            'target': j,
            'label': f"{source_label} - {target_label}",
            'weight': similarity_score
        }, ignore_index=True)

# Save the edges DataFrame to a CSV file
edges_file_path = '/content/drive/MyDrive/reddit_edges.csv'
edges_df.to_csv(edges_file_path, sep=';', index=False)
print(f"Edges DataFrame saved to: {edges_file_path}")
```
